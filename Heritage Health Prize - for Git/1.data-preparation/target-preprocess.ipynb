```python
import pandas as pd
import numpy as np
```

```python
df_dih_y2 = pd.read_csv('../data/DaysInHospital_Y2.csv')
df_dih_y3 = pd.read_csv('../data/DaysInHospital_Y3.csv')
```

```python
# Processing Y2 DIH Data
df_dih_y2['Year'] = 'Y1'
df_dih_y2['Year_actual'] = 'Y2'


# Processing Y3 DIH Data
df_dih_y3['Year'] = 'Y2'
df_dih_y3['Year_actual'] = 'Y3'
```

```python
# Union
df_dih = pd.concat([df_dih_y2, df_dih_y3]).reset_index().drop(['index'], axis=1)
```

```python
def count_nulls(df):
    null_counter = df.isnull().sum(axis=0)
    #null_counter = null_counter[null_counter > 0]
    null_percent = df.isnull().sum(axis=0) / df.shape[0] * 100
    #null_percent = null_percent[null_percent > 0]
    null_df = pd.concat([null_counter,null_percent],axis=1)
    null_df.columns = ['null_count','null_percent']
    display(null_df)
    
count_nulls(df_dih)
```

```python
# Export to csv for further processing
df_dih.to_csv("processed-data/target_dih.csv", index=False)
```
